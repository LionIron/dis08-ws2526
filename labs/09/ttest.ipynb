# Mein Dataset: Palmer Penguins

### Forschungsfrage (Hypothese)

Frage:

Haben Adelie-Pinguine und Gentoo-Pinguine ein unterschiedliches durchschnittliches Körpergewicht?

Nullhypothese (H₀)

Es gibt keinen Unterschied im mittleren Körpergewicht zwischen Adelie- und Gentoo-Pinguinen.

Alternativhypothese (H₁)

Es gibt einen signifikanten Unterschied im mittleren Körpergewicht.

# Wahl des Tests

zwei unabhängige Gruppen

metrische Variable (body_mass_g)

Vergleich von Mittelwerten

```python
import seaborn as sns
import pandas as pd
from scipy import stats

penguins = sns.load_dataset("penguins")

data = penguins[['species', 'body_mass_g']].dropna()

data.head()

```

```python
adelie_mass = data[data['species'] == 'Adelie']['body_mass_g']
gentoo_mass = data[data['species'] == 'Gentoo']['body_mass_g']

print("Adelie sample size:", len(adelie_mass))
print("Gentoo sample size:", len(gentoo_mass))
```

```python
print("Adelie mean body mass:", adelie_mass.mean())
print("Gentoo mean body mass:", gentoo_mass.mean())
```

Welch's t-test

```python
t_stat, p_value = stats.ttest_ind(
    adelie_mass,
    gentoo_mass,
    equal_var=False
)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis (statistically significant result).")
else:
    print("Fail to reject the null hypothesis (not statistically significant).")
```

Interpretation der Ergebnisse

Ein unabhängiger Zwei-Stichproben-t-Test wurde durchgeführt, um die Körpermasse von Adelie- und Gentoo-Pinguinen zu vergleichen.

Die Ergebnisse zeigen einen statistisch signifikanten Unterschied zwischen den beiden Gruppen
(p < 0,05). Daher wurde die Nullhypothese verworfen.

Gentoo-Pinguine haben im Durchschnitt eine deutlich höhere Körpermasse als Adelie-Pinguine. Dieses Ergebnis entspricht den biologischen Erwartungen und bestätigt, dass die Art ein starker Prädiktor für die Körpermasse ist.
