```python
import re

with open("./csv/orders.csv") as f:
    text = f.read()

print(text[:300])
```

```python
pattern = r'^([A-Z0-9]+),' \
          r'([A-Z0-9-]+),' \
          r'\$(\d+\.\d+),' \
          r'(\d+),' \
          r'(\d{2}/\d{2}/\d{4})$'

orders = re.findall(pattern, text, flags=re.MULTILINE)

len(orders), orders[0]
```

```python
order_numbers = [o[1] for o in orders]
order_numbers[:10]
```

```python
product_codes = [o[0] for o in orders]
product_codes[:10]
```

```python
prices = [float(o[2]) for o in orders]
prices[:10]
```

```python
dates = [o[4] for o in orders]
dates[:10]

```

```python
orders_over_500 = [o for o in orders if float(o[2]) > 500]
len(orders_over_500), orders_over_500[:5]
```

```python
converted_dates = [
    f"{d[3:5]}/{d[0:2]}/{d[6:]}" for d in dates
]
converted_dates[:10]
```

```python
items = [int(o[3]) for o in orders]
max_items = max(items)

orders_most_items = [o for o in orders if int(o[3]) == max_items]
max_items, orders_most_items
```

```python
min_price = min(prices)
cheapest_orders = [o for o in orders if float(o[2]) == min_price]

min_price, cheapest_orders
```
