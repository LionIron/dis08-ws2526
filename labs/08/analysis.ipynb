```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

penguins = sns.load_dataset("penguins")

penguins.head()
```

```python
penguins.info()
```

```python
penguins.isna().sum()
```

```python
penguins['species'].value_counts()
```

```python
penguins.describe()
```

```python
penguins.groupby('species').mean(numeric_only=True)
```

```python
plt.figure()
sns.histplot(data=penguins, x='body_mass_g', bins=30, kde=True)
plt.title("Distribution of Body Mass (g)")
plt.show()
```

```python
plt.figure()
sns.boxplot(data=penguins, x='species', y='body_mass_g')
plt.title("Body Mass by Penguin Species")
plt.show()
```

```python
plt.figure()
sns.scatterplot(
    data=penguins,
    x='flipper_length_mm',
    y='body_mass_g',
    hue='species'
)
plt.title("Flipper Length vs Body Mass")
plt.show()
```

```python
plt.figure()
sns.heatmap(
    penguins.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Matrix")
plt.show()
```

## Interpretation

- Gentoo penguins tend to have a higher body mass and longer flippers.
- There is a strong positive correlation between flipper length and body mass.
- Adelie penguins are generally smaller compared to the other species.
- The dataset shows clear separability between species, making it suitable
  for future classification tasks.
