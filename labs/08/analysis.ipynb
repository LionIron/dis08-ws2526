```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Style
sns.set(style="whitegrid")

# Load Palmer Penguins dataset
penguins = sns.load_dataset("penguins")

penguins.head()
```

```python
# Grundlegende Informationen
penguins.info()
```

```python
# Fehlende Werte
penguins.isna().sum()
```

```python
# Anzahl Beobachtungen pro Art
penguins['species'].value_counts()
```

```python
penguins.describe()
```

```python
penguins.groupby('species').mean(numeric_only=True)
```

```python
plt.figure()
sns.histplot(data=penguins, x='body_mass_g', bins=30, kde=True)
plt.title("Distribution of Body Mass (g)")
plt.show()
```

```python
plt.figure()
sns.boxplot(data=penguins, x='species', y='body_mass_g')
plt.title("Body Mass by Penguin Species")
plt.show()
```
